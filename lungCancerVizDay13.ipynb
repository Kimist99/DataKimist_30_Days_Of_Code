```python
#general list of packages for data viz in python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
#read in lung cancer data
df = pd.read_csv('dataLungCancer1.csv', encoding='latin-1')
df.head(3)
```

```python
#create bar chart looking at median tumor stage (T_stage = size of tumor) by age

tumor_size = df.groupby(by= 'T_stage').median()
tumor_size
```

```python
#display data with T_stage as column
tumor_size = df.groupby(by= 'T_stage').median().reset_index()
tumor_size
```

```python
#creat a bar plot of previous output usinf pandas graphing library
tumor_size.plot.bar(x="T_stage", y="age")
```

```python
#add semicolon to supress display about chart
#creat a bar plot of previous output
tumor_size.plot.bar(x="T_stage", y="age");
```

```python
# remove y axis to plot all data availabe in dataset that maps to x axis parameters
#creat a baR plot of previous output
tumor_size.plot.bar(x="T_stage");
```

```python
#matplotlib code for same plot
fig, ax = plt.subplots()

ax.bar(tumor_size["T_stage"], tumor_size["age"], label="age",
       
      edgecolor= "k", color=["red","blue","green","black"])# added to change color of each col

ax.set_xlabel("T_stage")

ax.legend();
```

```python
#use seaborn viz package to plot barchart
ax = sns.barplot("T_stage", "age", errcolor="black", capsize=0.1, data=df);
```

# Scatter Plots

```python
#using Pandas
df.plot.scatter("age", "tumorDose");
```

```python
#Using matplotlib
#fig = plt.figure(figsize=(20,10))
fig, ax = plt.subplots(figsize=(15,5)) #added figsize to increase size of graph
ax.scatter(df["age"], df["tumorDose"], marker="^",
           s=50, c=df["age"], edgecolors="k", alpha=0.5) #s = size of markers, c= color
ax.set_xlabel("age")
ax.set_ylabel("tumorDose");
plt.savefig("Lung Cancer Cohort Tumor Dose by Age2.jpeg")
```

```python
#using seaborn
plt.figure(figsize=(15,7))
p=sns.scatterplot("age","tumorDose", hue="age", s=30, edgecolor="none", data=df);
plt.savefig("Lung Cancer Cohort Tumor Dose by Age.jpeg") #save chart as jpeg
```
