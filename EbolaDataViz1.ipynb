```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
data = pd.read_csv('tableEbolaCDC.csv', encoding='latin-1')
data.head(3)
```

```python
df = data
df.head(10)
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=data, x='date', y='casesPerMo', sort=True)
plt.xticks(rotation=30)
plt.title('Total Counts of Ebola Cases, March 2014- April 2016')
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=data, x='date', y='casesPerMo')
plt.xticks(rotation=90)
plt.title('Total Counts of Ebola Cases, March 2014- April 2016')
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=data, x='Date', y='CasesPerMo')
plt.xticks(rotation=90)
plt.title('Total Counts of Ebola Cases, March 2014- April 2016')
```

```python
#add lables to bar chart
plt.figure(figsize=(20,10))
ax = sns.barplot(data=data, x='Date', y='CasesPerMo')

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=16, color='black', ha='center', va='bottom')

plt.xticks(rotation=90)
plt.title('Total Counts of Ebola Cases, March 2014- April 2016', fontsize=20)

#export viz as jpeg
plt.savefig("EbolaCountsMar14_Apr16.jpeg")
```
